```python
# Read the Example1.txt

companies1 = "companies.txt"
file1 = open(companies1, "r")
```

```python
# Close file after finish

file1.close()
```

### A Better Way to Open a File

```python
# Open file using with

with open(companies1, "r") as file1:
    FileContent = file1.read()
    
print(FileContent)
```

```python
# Verify if the file is closed

file1.closed
```

```python
# See the content of file

print(FileContent)
```

```python
# Read first four characters

with open(companies1, "r") as file1:
    print(file1.read(4))
```

```python
# Read certain amount of characters

with open(companies1, "r") as file1:
    print(file1.read(4))
    print(file1.read(4))
    print(file1.read(7))
    print(file1.read(15))
```

```python
# Read certain amount of characters

with open(companies1, "r") as file1:
     print(file1.read(16))
     print(file1.read(5))
     print(file1.read(9))
```

```python
# Read one line

with open(companies1, "r") as file1:
    print("first line: " + file1.readline())
```

```python
# Iterate through the lines

with open(companies1, "r") as file1:
    i = 0;
    for line in file1:
        print("Iteration", str(i), ": ", line)
        i = i + 1;
```

```python
# Read all lines and save as a list

with open(companies1, "r") as file1:
    FileasList = file1.readlines()
```
